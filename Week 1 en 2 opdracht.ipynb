## Data cleaning

De datasets worden eerst samengevoegd, tot een grote dataframe

Bij steps pakken we het gemiddelde en de mediaan. dit doen we nu al zodat die niet meer moet nadat ze zijn samengevoegd

```python
import seaborn as sns 
import pandas as pd 
import matplotlib.pyplot as plt 

df1 = pd.read_csv('survey.csv')
df1.head()
```

```python
df2 = pd.read_csv('steps.csv',sep=';')
df2['mean'] = df2.mean(axis=1)
df2['median'] = df2.median(axis=1)
df2.head()
```

```python
#De id variabele zit in beide grafieken dus deze gebruiken we om de dataset aan elkaar te binden

df = pd.merge(df1, df2, on = 'id')
df.head()
```

```python
df['weight'] = df['weight'].apply(pd.to_numeric, errors='coerce')

def delete(x):
    if(x > 200): 
        return float('NaN')
    elif(x <= 40):
        return float('NaN')
    else: 
        return x
    
df['weight'] = df['weight'].apply(delete)
```

```python
sns.distplot(df['weight'].dropna(), kde=True) 
plt.xlabel("Gewicht")
plt.show()
```
